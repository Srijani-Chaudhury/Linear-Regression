# Building Linear Regression Model from scratch


```python
class LinearRegression():
    def __init__(self):
        self.num=0
        self.base=0
        self.slope=0
        self.intercept=0
    def fit(self,data1,data2):
        mean1=np.mean(data1)
        mean2=np.mean(data2)
        
        for i in range(len(data1)):
            self.num += (data1[i]-mean1)*(data2[i]-mean2)
            self.base += (data1[i] - mean1)**2
       
        self.slope +=self.num/self.base
        self.intercept+=mean2-(self.slope*mean1)
        return self
    def predict(self,test_x):
        func=self.slope*test_x+self.intercept
        return func
    
    
```

# Testing 


```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x=data.iloc[:,0].values
y=data.iloc[:,1].values

trainx,testx,trainy,testy=train_test_split(x,y,test_size=0.2)

l=LinearRegression()
```

```python
l.fit(trainx,trainy)
y_pred = l.predict(testx)
y_pred
```

```python
print(r2_score(testy,y_pred))
```
