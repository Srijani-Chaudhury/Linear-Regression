# Building Linear Regression Model from scratch


```python
class LinearRegression():
    def __init__(self,lr=0.001):
        self.slope=0
        self.intercept=0
        self.lr=lr
    def fit(self,data1,data2):
        N=len(data1)
        for i in range(2000):
            Ypred= self.intercept + self.slope*data1
            s_b=(2/N)*sum(Ypred-data2)
            s_m=(2/N)*sum(data1*( Ypred-data2))
            
            self.slope = self.slope- (self.lr*s_m)
            self.intercept = self.intercept - (self.lr*s_b)
        return self
    def predict(self,test_x):
        func=self.slope*test_x+self.intercept
        return func
    
    
```

# Testing 


```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x=data.iloc[:,0].values
y=data.iloc[:,1].values

trainx,testx,trainy,testy=train_test_split(x,y,test_size=0.2)

l=LinearRegression()
```

```python
l.fit(trainx,trainy)
y_pred = l.predict(testx)
y_pred
```

```python
print(r2_score(testy,y_pred))
```
